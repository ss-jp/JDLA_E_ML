```python
#Mnistのインポート
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home="data")
```

```python
import numpy as np
import pandas as pd
data = mnist.data
data = data.astype(np.float64)
label = mnist.target
```

```python
#Mnistの正規化
data /= data.max()
```

```python
# mnistデータ数を格納
n = len(mnist.data)

# 実験に使うデータ数を格納
N=10000 

# ランダムにN枚を選択
indices = np.random.permutation(range(n))[:N] 

X = mnist.data[indices]
y = mnist.target[indices]
```

```python
Y = np.eye(10)[y.astype(int)] # 1-of-K表現に変換
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8)
```

<h1>Deep Learningモデル作成</h1>

```python
#import libralies

import tensorflow
from keras.layers.core import Activation, Flatten, Dense
import keras
```

```python
#モデル設定

n_in = len(X[0]) #784
n_hidden = 200
n_out=len(Y[0]) #10
```

```python
print('Building model...')
#modelの作成
model = keras.Sequential()

#入力から中間層の作成
model.add(Dense(n_hidden, input_dim=n_in))
model.add(Activation('sigmoid'))
model.add(Activation('sigmoid'))
model.add(Activation('sigmoid'))

#出力層の作成
model.add(Dense(n_out))
model.add(Activation('softmax'))
```

```python
#学習の設定
model.compile(loss='categorical_crossentropy',
                         optimizer=keras.optimizers.SGD(lr=0.007),
                         metrics=['accuracy'])
```

```python
#学習の実行
epochs=200
batch_size=200
history1 = model.fit(X_train, Y_train,  epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))
```

```python
#損失値の計算
score = model.evaluate(X_test, Y_test, verbose=1)
print('Test score (loss):', score[0])
print('Test accuracy:', score[1])
```

<h1>勾配消失問題の解決</h1>

```python
print('Building model...')
#modelの作成
relu_model = keras.Sequential()

#入力から中間層の作成
relu_model.add(Dense(n_hidden, input_dim=n_in))
relu_model.add(Activation('relu'))
relu_model.add(Activation('relu'))
relu_model.add(Activation('relu'))
relu_model.add(Activation('relu'))
# relu_model.add(Activation('relu'))
# relu_model.add(Activation('relu'))
# relu_model.add(Activation('relu'))

#出力層の作成
relu_model.add(Dense(n_out))
relu_model.add(Activation('softmax'))
```

```python
#学習の設定
relu_model.compile(loss='categorical_crossentropy',
                         optimizer=keras.optimizers.SGD(lr=0.01),
                         metrics=['accuracy'])
```

```python
#学習の実行
epochs=200
batch_size=200
history3 = relu_model.fit(X_train, Y_train,  epochs=epochs, batch_size=batch_size, validation_data=(X_test, Y_test))
```
